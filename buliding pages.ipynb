```python
<!DOCTYPE html>
<html>
<head>
    <title>Smart Lender</title>

    <style>
        body{
            font-family:Arial;
            background:#f5f5f5;
            text-align:center;
        }

        .container{
            width:70%;
            margin:auto;
            margin-top:80px;
            background:white;
            padding:40px;
            border-radius:10px;
            box-shadow:0px 0px 10px gray;
        }

        h1{
            color:#2E8B57;
        }

        p{
            font-size:18px;
        }

        button{
            background:#2E8B57;
            color:white;
            padding:15px 30px;
            border:none;
            border-radius:5px;
            font-size:18px;
            cursor:pointer;
        }

        button:hover{
            background:#1d6d45;
        }
    </style>

</head>

<body>

<div class="container">

<h1>Smart Lender</h1>

<h2>Loan Approval Prediction System</h2>

<p>

This application predicts whether a loan application will be approved based on applicant details using Machine Learning.

</p>

<br>

<a href="/predict">

<button>

Predict Loan Approval

</button>

</a>

</div>

</body>

</html>
<!DOCTYPE html>
<html>

<head>

<title>Loan Prediction</title>

<style>

body{

font-family:Arial;

background:#f4f4f4;

}

.container{

width:70%;

margin:auto;

background:white;

padding:20px;

margin-top:30px;

box-shadow:0px 0px 8px gray;

}

input,select{

width:100%;

padding:10px;

margin-bottom:15px;

}

input[type=submit]{

background:green;

color:white;

border:none;

cursor:pointer;

}

</style>

</head>

<body>

<div class="container">

<h2>Enter Applicant Details</h2>

<form action="/submit" method="POST">

<label>Gender</label>

<select name="Gender">

<option value="0">Male</option>

<option value="1">Female</option>

</select>

<label>Married</label>

<select name="Married">

<option value="1">Yes</option>

<option value="0">No</option>

</select>

<label>Dependents</label>

<input type="number" name="Dependents">

<label>Education</label>

<select name="Education">

<option value="0">Graduate</option>

<option value="1">Not Graduate</option>

</select>

<label>Self Employed</label>

<select name="Self_Employed">

<option value="1">Yes</option>

<option value="0">No</option>

</select>

<label>Applicant Income</label>

<input type="number" name="ApplicantIncome">

<label>Co Applicant Income</label>

<input type="number" name="CoapplicantIncome">

<label>Loan Amount</label>

<input type="number" name="LoanAmount">

<label>Loan Amount Term</label>

<input type="number" name="Loan_Amount_Term">

<label>Credit History</label>

<input type="number" name="Credit_History">

<label>Property Area</label>

<select name="Property_Area">

<option value="0">Urban</option>

<option value="1">Semiurban</option>

<option value="2">Rural</option>

</select>

<input type="submit" value="Predict">

</form>

</div>

</body>

</html>
<!DOCTYPE html>

<html>

<head>

<title>Prediction Result</title>

<style>

body{

background:#f4f4f4;

font-family:Arial;

text-align:center;

}

.container{

width:60%;

margin:auto;

margin-top:100px;

background:white;

padding:40px;

box-shadow:0px 0px 8px gray;

}

h1{

color:green;

}

button{

padding:12px 25px;

background:green;

color:white;

border:none;

cursor:pointer;

}

</style>

</head>

<body>

<div class="container">

<h1>{{ prediction_text }}</h1>

<br>

<a href="/predict">

<button>

Predict Another Loan

</button>

</a>

</div>

</body>

</html>
from flask import Flask, render_template, request
import pickle
import numpy as np

app = Flask(__name__)

model = pickle.load(open("rdf.pkl","rb"))
scaler = pickle.load(open("scale1.pkl","rb"))

@app.route("/")
def home():
    return render_template("home.html")

@app.route("/predict")
def predict():
    return render_template("predict.html")

@app.route("/submit", methods=["POST"])
def submit():

    values = [float(x) for x in request.form.values()]

    final = scaler.transform([values])

    prediction = model.predict(final)

    if prediction[0] == 1:
        result = "Loan Approved"
    else:
        result = "Loan Rejected"

    return render_template("submit.html",
                           prediction_text=result)

if __name__ == "__main__":
    app.run(debug=True)
    python app.py
    http://127.0.0.1:5000
    Home Page
      │
      ▼
Click "Predict Loan Approval"
      │
      ▼
Predict Page (Enter Details)
      │
      ▼
Click Submit
      │
      ▼
Model Predicts
      │
      ▼
Submit Page
(Loan Approved / Loan Rejected)
templates/
    home.html
    input.html
    output.html
    
```
